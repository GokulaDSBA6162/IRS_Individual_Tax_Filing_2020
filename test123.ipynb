```python
import altair as alt
freq_chart = alt.Chart(state_agg).mark_bar().transform_filter(
            alt.FieldGTEPredicate(field='Total income in Amount', gte= 10000)
            ).encode(
        y=alt.Y('State',sort=alt.EncodingSortField('Total income in Amount', op='min', order='descending')),
        x='Total income in Amount'
    )



freq_text = freq_chart.mark_text(
    align='left',
    baseline='middle',
    dx=3
).encode(
    text=alt.Text('Total income in Amount',format='.1f')
)
freq_chart + freq_text

```
